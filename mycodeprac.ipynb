```python
#Binary search
def binary_search(arr,x):
    l,r = 0, len (arr)-1
    while l <= r:
        mid = (l+r)//2
        if arr[mid] == x:
            return mid
        elif arr[mid] < x:
            l = mid+1
        else:
            r =  mid-1
            return -1
```

```python
#Linear search
def linear(arr,x):
    for i in range(len(arr)):
        if arr[i] ==x :
            return i
```

```python
#Bubble sort
arr = [5,3,1]
for i in range(len(arr)):
    for j in range(len(arr)-1):
        if arr[j] > arr[j+1]:
            arr[j],arr[j+1] = arr[j+1], arr[j]
print(arr)
```
